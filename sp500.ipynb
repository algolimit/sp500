```python
eventsdf = pd.read_csv("Economic_Events.csv")
eventsdf
```

```python
pd.to_datetime(eventsdf["Date"], errors="raise", yearfirst=True)
eventsdf
```

```python
eventsdf = eventsdf.drop(columns="Impact on Global Markets")
eventsdf
```

```python
# We willen de Ukrainse oorlog handmatig toevoegen
ukraine_war_event = {
    'Date': ['2022-02-24'],
    'Event': ['Ukraine War'],
    'Impact on S&P 500': ['Negative - Increased volatility and inflation concerns'],
    'Remarks': ['Has led to significant disruptions in global commodity markets, including energy and grains, influencing global inflation rates and market stability.'],
    'Global Market Impact Score': [7.0]
}

new_row = pd.DataFrame(ukraine_war_event)

eventsdf = pd.concat([eventsdf, new_row], ignore_index=True)
eventsdf = eventsdf.sort_values("Date", ignore_index=True)
eventsdf
```

```python
sp500df.to_csv("SP500_Dataset")
eventsdf.to_csv("Events_Dataset")
```
