```python
stockmarkets_df = pd.read_excel("Stock Markets, US$.xlsx")

stockmarkets_df = stockmarkets_df.drop(stockmarkets_df.index[0])

stockmarkets_df["Unnamed: 0"] = stockmarkets_df["Unnamed: 0"].astype(int)

stockmarkets_df["Unnamed: 0"] = pd.to_datetime(stockmarkets_df["Unnamed: 0"], format="%Y")

stockmarkets_df = stockmarkets_df[["Unnamed: 0", "High Income Countries"]]

stockmarkets_df["Date"] = stockmarkets_df["Unnamed: 0"]

stockmarkets_df = stockmarkets_df[["Date", "High Income Countries"]]

stockmarkets_df = stockmarkets_df.rename(columns={"High Income Countries": "High Income Countries (SM)"})

stockmarkets_df
```

```python
GDP_df = pd.read_csv("API_NY.GDP.MKTP.CD_DS2_en_csv_v2_31795/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_31795.csv")

GDP_df = GDP_df[GDP_df["1960"] > 13263458995.00]

GDP_df
```

```python
GDP_df = GDP_df.drop(columns={"Country Code", "Indicator Name", "Indicator Code", "Unnamed: 68"})

GDP_df
```

```python
melted_df = pd.melt(GDP_df, id_vars=['Country Name'], var_name='Year', value_name='Value')

reshaped_df = melted_df.pivot(index='Year', columns='Country Name', values='Value')

GDP_df = reshaped_df[["World"]]

GDP_df = GDP_df.rename(columns={"World": "World GDP"})

GDP_df_reset = GDP_df.reset_index()

GDP_df_reset.rename(columns={'Year': 'Date'}, inplace=True)

GDP_df = GDP_df_reset
GDP_df_reset = GDP_df.reset_index(drop=True)  

GDP_df_reset.columns = ['Date', 'World GDP']

GDP_df_reset.index.name = None

GDP_df = GDP_df_reset

GDP_df["Date"] = pd.to_datetime(GDP_df["Date"])

GDP_df
```

```python
stats_df = pd.merge(unemploy_df, stockmarkets_df, on="Date", how="inner")

stats_df = pd.merge(stats_df, GDP_df, on="Date", how="inner")

stats_df
```
